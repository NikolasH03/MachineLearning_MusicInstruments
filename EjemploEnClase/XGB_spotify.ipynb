La sintaxis usada en XGBoost es muy similar a la usada al momento de implementar modelos de Scikit-Learn.

Entrenamiento con parada temprana ("early stopping")

XGBoost incluye, entre otras opciones, el entrenamiento con parada temprana.

La idea básica es simple: se monitorea el desempeño del modelo durante el entrenamiento y sobre el set de validación y si el desempeño no mejora durante cierto número de iteraciones ("boosting rounds") entonces se detiene el entrenamiento.

Supongamos que queremos detener el entrenamiento si después de 10 iteraciones el desempeño no mejora. Para ello: usaremos un número elevado de árboles como hiper-parámetro del modelo (por ejemplo 1.000)

Definiremos el hiper-parámetro "early_stopping_rounds" con un valor de 10 (que serán las iteraciones a tener en cuenta para detener o continuar con el entrenamiento)

No importa si usamos un número elevado de árboles, pues probablemente el entrenamiento se detendrá antes de construir la totalidad de estos árboles.

```python
# Definir parámetros incluyendo early-stopping
#Desactivamos la metrica de evaluación
param = {
    'booster': 'gbtree',  #árboles para xgboost
    'eta': 0.1, #tasa de aprendizaje
    'gamma': 0,
    'lambda': 0,
    'max_depth': 6,
    'colsamples_bytree': 0.8,   #porcentaje de caracteristicas a usar
    'objective': 'multi:softmax',
    'n_estimators': 1000,  #cantidad de estimadores o árboles  ***
    'random_state': 42,
    'early_stopping_rounds': 10,  #número de iteraciones de parada temprana
    'num_class': 6,
    'enable_categorical': False,
    'eval_metric': 'merror',  #métrica de evaluación m porque es multiclase  ***
}
```

```python
# Crear instancia del modelo con los hiper-parámetros especificados

modelo = XGBClassifier(**param)
modelo.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=True)
```

a pesar de haber definido un total de 1.000 árboles, el entrenamiento se detiene cuando el modelo alcanza aproximadamente 103 árboles.

En el caso anterior, la métrica de desempeño de referencia usada para el "early stopping" fue la configurada para multiclase "merror", se calcula como .#(wrong cases)/#(all cases)

Ahora se hará con: mlogloss



```python
# Definir parámetros incluyendo early-stopping
#Desactivamos la metrica de evaluación
param = {
    'booster': 'gbtree',  #árboles para xgboost
    'eta': 0.1, #tasa de aprendizaje
    'gamma': 0,
    'lambda': 0,
    'max_depth': 6,
    'colsamples_bytree': 0.8,   #porcentaje de caracteristicas a usar
    'objective': 'multi:softmax',
    'n_estimators': 1000,  #cantidad de estimadores o árboles  ***
    'random_state': 42,
    'early_stopping_rounds': 10,  #número de iteraciones de parada temprana
    'num_class': 6,
    'enable_categorical': False,
    'eval_metric': 'mlogloss',  #métrica de evaluación m porque es multiclase  ***
}
```

```python
# Crear instancia del modelo con los hiper-parámetros especificados

modelo = XGBClassifier(**param)
modelo.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=True)
```

Y podemos usar el "early-stopping" como parte del proceso de validación cruzada:

```python
# Convertir el set de entrenamiento al formato DMatrix usado por XGBoost
dtrain2 = xgb.DMatrix(data=X_train, label=y_train, enable_categorical=False)

# Hiper-Parámetros
param2 = {
    'booster': 'gbtree', # Usar gradient boosting trees
    'eta': 0.1, # Tasa de aprendizaje
    'gamma': 0, # Parámetro 𝜸
    'lambda': 0, # Parámetro λ
    'max_depth': 6, # Máxima profundidad del árbol
    'colsample_bytree': 0.8, # Porcentaje de muestreo de las columnas
    'num_class': 6,
    'objective': 'multi:softprob',  #SE CAMBIO EL OBJETIVO:  El resultado contiene la probabilidad predicha de que cada punto de datos pertenezca a cada clase
    'random_state': 42, # Semilla del generador de números aleatorios
    'eval_metric': 'merror', # Usar tasa de error (= 1 - exactitud) como métrica de desempeño
}
```

```python

# Validación cruzada (5-fold) (no es necesario definir el tipo de modelo, este se infiere del hiper-parámetro "objective")
resultado2_CV = xgb.cv(
    params=param2,
    dtrain=dtrain,
    num_boost_round=1000,           # Definir inicialmente un número elevado de árboles
    nfold=5,                        # Número de particiones
    seed=123,                       # Semilla generador aleatorio
    early_stopping_rounds=10,       # **** HIPER-PARÁMETROS EARLY STOPPING ****
    as_pandas=True                  # Retornar resultados como DataFrame de Pandas
)

resultado2_CV
```

Y vemos que tras entrenar 57 árboles el entrenamiento se detiene. En este punto podemos imprimir en pantalla el desempeño obtenido con entrenamiento/validación, donde el mejor desempeño corresponderá a la última iteración:

```python
mejor_ex_train = 1 - resultado2_CV['train-merror-mean'].iloc[-1]
mejor_ex_test = 1 - resultado2_CV['test-merror-mean'].iloc[-1]

# Imprimir resultados en pantalla
print(f'Mejor exactitud entrenamiento: {100*mejor_ex_train:.1f}%')
print(f'Mejor exactitud prueba: {100*mejor_ex_test:.1f}%')
```
